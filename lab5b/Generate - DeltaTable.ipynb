```python
import random

# ==========================================
# 1. CONFIGURACIÓN DE ADLS GEN2
# ==========================================
storage_account_name = "TU_CUENTA_DE_ALMACENAMIENTO"
container_name = "TU_CONTENEDOR"
access_key = "TU_ACCESS_KEY_AQUI"

# Ruta ABFSS hacia la capa Gold
output_path = f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net/gold/iot_sensors"

# ==========================================
# 2. AUTENTICACIÓN MEDIANTE ACCESS KEY
# ==========================================
spark.conf.set(
    f"fs.azure.account.key.{storage_account_name}.dfs.core.windows.net",
    access_key
)

# ==========================================
# 3. GENERAR DATOS SIMULADOS (100 SENSORES)
# ==========================================
# Creamos una lista vacía y la llenamos con un bucle
data = []
estados_posibles = ["Activo", "Activo", "Activo", "Inactivo", "Mantenimiento"]

for i in range(1, 101): # Genera del 1 al 100
    id_sensor = i
    dispositivo = f"Sensor_{i:03d}" # Formato: Sensor_001, Sensor_015, etc.
    
    # Simulamos que los inactivos/mantenimiento leen 0.0, el resto un valor aleatorio
    estado = random.choice(estados_posibles)
    if estado == "Activo":
        lectura = round(random.uniform(10.0, 450.0), 2)
    else:
        lectura = 0.0
        
    data.append((id_sensor, dispositivo, lectura, estado))

columns = ["id", "dispositivo", "lectura_ppm", "estado"]

# Convertimos la lista generada a un DataFrame de Spark
df = spark.createDataFrame(data, columns)

print(f"Se han generado {df.count()} registros de prueba.")
display(df) 

# ==========================================
# 4. ESCRIBIR EN FORMATO DELTA
# ==========================================
print(f"Escribiendo datos Delta en: {output_path} ...")

df.write \
    .format("delta") \
    .mode("overwrite") \
    .save(output_path)

print("¡Proceso finalizado! Se ha generado la tabla Delta y el _delta_log con 100 sensores.")
```
